```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
url = "http://wiki.lspace.org/mediawiki/List_of_Pratchett_characters"

def getSoup(url):
    r = requests.get(url)
    r.encoding = 'UTF-8'
    return BeautifulSoup(r.text,'lxml')

def getLinks(soup):
    lis = soup.find('span',{'id':'Unseen_University_and_the_Wizards'}).parent.previous.previous.parent.parent.parent.find_all('li')
    
    links = []
    for l in lis:
        link = "http://wiki.lspace.org" + l.find('a')['href']
        links.append(link)
         
    return links

def getChars(link):
    soup = getSoup(link)
    name = soup.find(id='firstHeading').text
    
    books = [i.text for i in soup.select("td a[href*='Book:']")]
    
    chars = {
        'Name': name,
        'Books': books
    }
    
    
    return chars


soup = getSoup(url)
links = getLinks(soup)
names = [getChars(link) for link in links[:2]]
names


rows = [] 
  
for data in names: 
    books = data['Books'] 
    name = data['Name'] 
    
    for book in books:
        row = {'Name':name, 'Book':book}
        rows.append(row)
        
pd.DataFrame(rows)
  




```
